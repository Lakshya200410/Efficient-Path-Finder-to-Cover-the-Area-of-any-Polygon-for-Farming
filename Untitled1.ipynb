```python
# Farm boundary (irregular shape)
farm_boundary = Polygon([
    (0, 0), (15, 0), (20, 5), (18, 8), (12, 10), (8, 7), (4, 12), (1, 8), (0, 4)
])

# Circular obstacle
circle_center = Point(10, 6)
radius = 2
circular_obstacle = circle_center.buffer(radius).simplify(0.1)  # Approximate circle

# List of obstacles
obstacles = [circular_obstacle]

# Function to generate a coverage path
def generate_coverage_path(farm, equipment_width, obstacles=[]):
    # Create a bounding box around the farm
    minx, miny, maxx, maxy = farm.bounds

    # Generate parallel lines spaced by equipment width
    lines = []
    y = miny
    direction = 1  # Alternate direction to create boustrophedon pattern
    while y <= maxy:
        line = LineString([(minx, y), (maxx, y)])
        line = line.intersection(farm)  # Clip to farm boundary

        # Remove line sections overlapping obstacles
        for obs in obstacles:
            line = line.difference(obs)

        # Handle MultiLineString or empty cases
        if line.is_empty:
            y += equipment_width
            continue
        if isinstance(line, LineString):
            # Reverse direction for boustrophedon pattern
            if direction == -1:
                line = LineString(reversed(line.coords))
            lines.append(line)
        elif isinstance(line, MultiLineString):
            for subline in line.geoms:  # Use .geoms to access individual LineStrings
                if direction == -1:
                    subline = LineString(reversed(subline.coords))
                lines.append(subline)
        y += equipment_width
        direction *= -1  # Flip direction for the next line

    return lines


# Function to plot the coverage path
def plot_coverage_path(farm, obstacles, coverage_path):
    x, y = farm.exterior.xy
    plt.plot(x, y, label='Farm Boundary', color='blue')
    for obs in obstacles:
        ox, oy = obs.exterior.xy
        plt.fill(ox, oy, color='red', alpha=0.5, label='Obstacle')
    for line in coverage_path:
        if isinstance(line, LineString):
            lx, ly = line.xy
            plt.plot(lx, ly, color='green')
    plt.legend()
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Coverage Path")
    plt.grid(True)
    plt.show()

# Generate coverage path
equipment_width = 1.0  # Equipment coverage width (1 meter)
coverage_path = generate_coverage_path(farm_boundary, equipment_width, obstacles)

# Plot the coverage path
plot_coverage_path(farm_boundary, obstacles, coverage_path)
```
